```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor


CSV_PATH = "dataset/clean_car_dataset.csv"

df = pd.read_csv(CSV_PATH)
print(df.head())

# x = df.drop(columns = ["Price","LogPrice"])
x = df.drop(columns=["Unnamed: 0", "Price", "LogPrice", "Year"])
y = df["LogPrice"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)

lr_pred= lr.predict(x_test)
```

```python

rf = RandomForestRegressor(
    n_estimators=100, 
    max_depth=5,            
    min_samples_leaf=5,      
    random_state=42
)
rf.fit(x_train, y_train)



gbr = GradientBoostingRegressor(
    n_estimators=150, 
    learning_rate=0.05,
    max_depth=3,           
    random_state=42
)

gbr.fit(x_train, y_train)


gbr_pred_log = gbr.predict(x_test)
gbr_pred_actual = np.exp(gbr_pred_log)


lr_pred_log = lr.predict(x_test)
rf_pred_log = rf.predict(x_test)

lr_pred = np.exp(lr_pred_log)
rf_pred = np.exp(rf_pred_log)
y_test_actual= np.exp(y_test)

```

```python
print("prediction linear regration")
print(lr_pred)
rf_pred = rf.predict(x_test)
print("prediction random forest")
print(rf_pred)
print("prediction xgboost")
```

```python
def print_matrix(name,y_true,y_pred):
    r2 = r2_score(y_true,y_pred)
    mea = mean_absolute_error(y_true,y_pred)
    mse = mean_squared_error(y_true,y_pred)
    rmse = np.sqrt(mse)


    print(f"\n {name} performence")
    print(f"r2 {r2:.3f}")
    print(f"mea {mea:.0f}")    
    print(f"mse {mse:.0f}")    
    print(f"rms {rmse:.0f}")
lr_pred_actual = np.exp(lr.predict(x_test))
rf_pred_actual = np.exp(rf.predict(x_test))
y_test_actual = np.exp(y_test)

print_matrix("Linear Regression Performance", y_test_actual, lr_pred_actual)
print_matrix("Random Forest Performance", y_test_actual, rf_pred_actual)
print_matrix("Gradient Boosting: ", y_test_actual, gbr_pred_actual)
```

```python
i = 20
x_one_df = x_test.iloc[[i]]

y_true_actual = np.exp(y_test.iloc[i])

p_lr_one_log = float(lr.predict(x_one_df)[0])
p_rf_one_log = float(rf.predict(x_one_df)[0])
p_gbr_one_log = float(gbr.predict(x_one_df)[0])

p_lr_one_actual = np.exp(p_lr_one_log)
p_rf_one_actual = np.exp(p_rf_one_log)
p_gbr_one_actual = np.exp(p_gbr_one_log)


print(f"Actual Price     : ${y_true_actual:,.0f}")
print(f"Linear Regression: ${p_lr_one_actual:,.0f}")
print(f"Random Forest    : ${p_rf_one_actual:,.0f}")
print(f"Gradient Boosting: ${p_gbr_one_actual:,.0f}")
```

```python
print("Linear Train :", lr.score(x_train, y_train))
print("Linear Test  :", lr.score(x_test, y_test))

print("RF Train :", rf.score(x_train, y_train))
print("RF Test  :", rf.score(x_test, y_test))

print("GB Train :", gbr.score(x_train, y_train))
print("GB Test  :", gbr.score(x_test, y_test))
```
