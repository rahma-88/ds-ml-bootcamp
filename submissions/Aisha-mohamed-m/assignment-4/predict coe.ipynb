```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import  RandomForestRegressor
from sklearn.metrics import  mean_absolute_error, mean_squared_error, r2_score
print(np.__version__)
```

```python
CSV_PATH = "cars_clean.csv"
df = pd.read_csv(CSV_PATH)
print(df.head)
```

```python
# 2. Qeex target-ka (y) iyo sifooyinka (X)
# Waxaan bartilmaameed ka dhignay 'logPrice' si aan u helno saadaal dhab ah oo kala duwan
y = df['logPrice']
x = df.drop(columns=['logPrice', 'Location_AVg_Price', 'Unnamed: 0'], errors='ignore')

# 3. Qaybi xogta (Train/Test Split)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr =LinearRegression()
lr.fit(x_train,y_train )
lr_predict = lr.predict(x_test )
print(lr_predict)
```

```python
rf =RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train,y_train )
rf_predict = rf.predict(x_test )
print(rf_predict )
```

```python
# 1. Qeex function-ka (Dhammaan print-yadu waa inay afar firaaqo/spaces ka dambeeyaan geesta)
def print_metrics(name, y_true, y_pred):
    y_true_real = np.exp(y_true)
    y_pred_real = np.exp(y_pred)
    
    r2 = r2_score(y_true_real, y_pred_real)
    mae = mean_absolute_error(y_true_real, y_pred_real)
    mse = mean_squared_error(y_true_real, y_pred_real)
    rmse = np.sqrt(mse)
    
    print(f"\n{name} performance:")
    print(f"R-squared: {r2:.3f}")
    print(f"MAE: ${mae:,.0f}")
    print(f"MSE: {mse:,.0f}")
    print(f"RMSE: ${rmse:,.0f}")

# 2. U wac labada moodel (Kuwani waxay ku noqonayaan geesta bidix ee rasmiga ah)
print_metrics("LinearRegression", y_test, lr_predict)
print_metrics("RandomForest", y_test, rf_predict)
```

```python
i = 5 # Waxaad u beddeli kartaa nambar kasta (tusaale: 1, 2, 3...)
x_one_df = x_test.iloc[[i]]

# Qiimayaasha log-ga ahaa waxaan ku celinaynaa doolar dhab ah ($)
y_true_one_row = np.exp(y_test.iloc[i])
p_lr_one = np.exp(float(lr.predict(x_one_df)[0]))
p_rf_one = np.exp(float(rf.predict(x_one_df)[0]))

print(f"Prediction for row index {i}:")
print(f"Actual price is: ${y_true_one_row:,.0f}")
print(f"Predicted LR price is: ${p_lr_one:,.0f}")
print(f"Predicted RF price is: ${p_rf_one:,.0f}")
```
