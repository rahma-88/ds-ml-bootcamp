# Loan Approval Prediction

Models Used:

1. Logistic Regression
2. Random Forest
3. Decision Tree

```python
# =====================================
# Assignment Five Part B2
# Loan Classification
# =====================================

import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import RandomForestClassifier

from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import *


df=pd.read_csv(
"clean_loan_dataset.csv"
)


y=df["Approved"]

X=df.drop(
"Approved",
axis=1
)


X_train,X_test,y_train,y_test=train_test_split(

X,
y,
test_size=.20,
stratify=y,
random_state=42

)


logistic=LogisticRegression(

max_iter=1000,
random_state=42
)


forest=RandomForestClassifier(

n_estimators=100,
random_state=42
)


tree=DecisionTreeClassifier(

random_state=42
)


logistic.fit(
X_train,
y_train
)

forest.fit(
X_train,
y_train
)

tree.fit(
X_train,
y_train
)



def evaluate(model,name):


    pred=model.predict(
    X_test
    )


    print("\n"+name)

    print(
    "Accuracy:",
    round(
    accuracy_score(
    y_test,
    pred
    ),3)
    )

    print(
    "Precision:",
    round(
    precision_score(
    y_test,
    pred
    ),3)
    )


    print(
    "Recall:",
    round(
    recall_score(
    y_test,
    pred
    ),3)
    )


    print(
    "F1:",
    round(
    f1_score(
    y_test,
    pred
    ),3)
    )


    print(
    "\nConfusion Matrix"
    )

    print(
    confusion_matrix(
    y_test,
    pred
    )
    )



evaluate(
logistic,
"Logistic Regression Performance"
)

evaluate(
forest,
"Random Forest Performance"
)

evaluate(
tree,
"Decision Tree Performance"
)



# Sanity Check

sample=X_test.iloc[[0]]

print("\nActual")

print(
y_test.iloc[0]
)


print(
"\nLogistic:",
logistic.predict(sample)[0]
)


print(
"\nRandom Forest:",
forest.predict(sample)[0]
)


print(
"\nDecision Tree:",
tree.predict(sample)[0]
)
```
