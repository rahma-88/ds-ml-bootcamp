```python
# ==========================================
# STEP 1 : Import Required Libraries
# ==========================================

# Import pandas for data manipulation
import pandas as pd

# Import numpy for numerical operations
import numpy as np

# Import SimpleImputer for handling missing values
from sklearn.impute import SimpleImputer

# Import RobustScaler for feature scaling
# (Chosen instead of StandardScaler)
from sklearn.preprocessing import RobustScaler

# Display all dataframe columns
pd.set_option("display.max_columns", None)

```

```python
# ==========================================
# STEP 2 : Load Dataset
# ==========================================

# Read the CSV file

df = pd.read_csv("/content/raw_loan_dataset.csv")

# Print confirmation

print("=" * 60)
print("Dataset loaded successfully.")
print("=" * 60)

# Display dataset dimensions

print("Number of Rows :", df.shape[0])
print("Number of Columns :", df.shape[1])
```

```python
# ==========================================
# STEP 3 : Inspect Dataset
# ==========================================

# Display first five rows

print("\nFirst Five Rows")
display(df.head())

# Display dataset information

print("\nDataset Information")
print("-" * 40)

df.info()

# Display missing values

print("\nMissing Values")
print("-" * 40)

print(df.isnull().sum())

# Display statistical summary

print("\nSummary Statistics")
print("-" * 40)

display(df.describe(include="all"))

```

```python
# ==========================================
# STEP 4 : Identify Problems
# ==========================================

# Display sample values from Income

print("\nSample values from Income")
print("-" * 40)

print(df["Income"].head(10))

# Display sample values from LoanAmount

print("\nSample values from LoanAmount")
print("-" * 40)

print(df["LoanAmount"].head(10))

# Check unique values in HasCollateral

print("\nUnique values in HasCollateral")
print("-" * 40)

print(df["HasCollateral"].value_counts(dropna=False))

# Check unique values in PreviousDefaults

print("\nUnique values in PreviousDefaults")
print("-" * 40)

print(df["PreviousDefaults"].value_counts(dropna=False))

# Check unique values in Approved

print("\nUnique values in Approved")
print("-" * 40)

print(df["Approved"].value_counts(dropna=False))

```

```python
# ==========================================
# STEP 5 : Clean Currency Formatting
# ==========================================

# Display data types before cleaning
print("=" * 60)
print("Data Types Before Cleaning")
print("=" * 60)

print(df[["Income", "LoanAmount"]].dtypes)


# List of currency columns
currency_columns = ["Income", "LoanAmount"]


# Loop through each currency column
for column in currency_columns:

    # Convert values to string first
    df[column] = df[column].astype(str)

    # Remove dollar sign ($)
    df[column] = df[column].str.replace("$", "", regex=False)

    # Remove commas (,)
    df[column] = df[column].str.replace(",", "", regex=False)

    # Convert cleaned values into numeric values
    # Invalid values become NaN
    df[column] = pd.to_numeric(df[column], errors="coerce")


print("\nCurrency formatting cleaned successfully.")
print("=" * 60)

# Display data types after cleaning
print(df[["Income", "LoanAmount"]].dtypes)

print("\nSample cleaned values")
display(df[["Income", "LoanAmount"]].head())
```

```python
# ==========================================
# STEP 6 : Fix Category Errors
# ==========================================

# Dictionary used to normalize all category values
category_mapping = {

    # Yes values
    "Yes": "Yes",
    "YES": "Yes",
    "yes": "Yes",
    "Y": "Yes",
    "y": "Yes",
    "Approved": "Yes",
    "approved": "Yes",

    # No values
    "No": "No",
    "NO": "No",
    "no": "No",
    "N": "No",
    "n": "No",
    "Rejected": "No",
    "rejected": "No"

}


# Columns that require normalization
category_columns = [

    "HasCollateral",
    "PreviousDefaults",
    "Approved"

]


# Loop through every categorical column
for column in category_columns:

    print("=" * 60)
    print(f"Cleaning {column}")
    print("=" * 60)

    # Replace inconsistent values
    df[column] = df[column].replace(category_mapping)

    # Display value counts after cleaning
    print(df[column].value_counts(dropna=False))
```

```python
# ==========================================
# STEP 7 : Handle Missing Values
# ==========================================

# Identify numerical columns
numeric_columns = df.select_dtypes(include=np.number).columns

# Identify categorical columns
categorical_columns = df.select_dtypes(include="object").columns


# Create median imputer
numeric_imputer = SimpleImputer(strategy="median")


# Create mode imputer
categorical_imputer = SimpleImputer(strategy="most_frequent")


# Fill missing numerical values
df[numeric_columns] = numeric_imputer.fit_transform(df[numeric_columns])


# Fill missing categorical values
df[categorical_columns] = categorical_imputer.fit_transform(df[categorical_columns])


print("=" * 60)
print("Missing values after imputation")
print("=" * 60)

print(df.isnull().sum())
```

```python
# ==========================================
# STEP 8 : Remove Duplicate Rows
# ==========================================

# Count rows before removing duplicates
rows_before = len(df)

print("=" * 60)
print("Rows Before Removing Duplicates")
print(rows_before)


# Remove duplicate rows
df = df.drop_duplicates()


# Count rows after removing duplicates
rows_after = len(df)

print("\nRows After Removing Duplicates")
print(rows_after)

print("\nDuplicate Rows Removed")
print(rows_before - rows_after)

print("=" * 60)
print("Duplicate removal completed.")
print("=" * 60)
```

```python
# ==========================================
# STEP 9 : Handle Outliers Using IQR Capping
# ==========================================

# List of numerical columns for outlier treatment
outlier_columns = [
    "Income",
    "CreditScore",
    "LoanAmount",
    "EmploymentYears"
]

# Loop through each numerical column
for column in outlier_columns:

    # Calculate first quartile (25%)
    Q1 = df[column].quantile(0.25)

    # Calculate third quartile (75%)
    Q3 = df[column].quantile(0.75)

    # Calculate Interquartile Range
    IQR = Q3 - Q1

    # Calculate lower and upper bounds
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    # Cap extreme values instead of removing rows
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)

```

```python
# ==========================================
# STEP 10 : Label Encoding
# ==========================================

# Dictionary for encoding
label_mapping = {
    "Yes": 1,
    "No": 0
}

# Columns to encode
label_columns = [
    "Approved",
    "HasCollateral",
    "PreviousDefaults"
]

# Convert Yes/No into 1/0
for column in label_columns:
    # Check if there are any NaNs in the column.
    if df[column].isnull().any():
        modes = df[column].mode()
        if not modes.empty:
            # If mode(s) exist, fill NaNs with the most frequent value
            mode_val = modes.iloc[0]
            df[column] = df[column].fillna(mode_val)
        else:
            # If modes is empty, the column is entirely NaN.
            # In this case, we need to decide on a default fill value before mapping.
            # For 'Approved', 'HasCollateral', 'PreviousDefaults', 'No' (which maps to 0)
            # is a reasonable default if the column was completely missing.
            print(f"Warning: Column '{column}' was entirely NaN. Filling with 'No' as default for mapping.")
            df[column] = df[column].fillna('No') # Fill with string 'No'

    # Ensure values are string, strip whitespace, then map
    df[column] = df[column].astype(str).str.strip().map(label_mapping)

    print(f"{column} encoded successfully.")

print("=" * 60)
print("Class Distribution")
print("=" * 60)

# Display class distribution
print(df["Approved"].value_counts())
```

```python
# ==========================================
# STEP 11 : Class Balance Check
# ==========================================

print("=" * 60)
print("Approved vs Rejected Counts")
print("=" * 60)

print(df["Approved"].value_counts())

print()

print("=" * 60)
print("Class Proportions")
print("=" * 60)

print(df["Approved"].value_counts(normalize=True))

print()

# Simple interpretation
ratio = df["Approved"].value_counts(normalize=True)

if ratio.empty:
    print("The 'Approved' column does not contain any valid values for classification balance check.")
    print("This indicates a potential data integrity issue or an empty dataset for this column.")
    print("Cannot perform class balance check.")
elif len(ratio) == 1:
    print("The 'Approved' column contains only one unique class after encoding.")
    print("This indicates an extremely imbalanced dataset where only one outcome exists.")
    print("Precision, Recall, and F1-score are necessary evaluation metrics.")
else: # len(ratio) >= 2, assuming binary classification for this check
    if abs(ratio.iloc[0] - ratio.iloc[1]) < 0.20:
        print("The dataset is reasonably balanced.")
        print("Accuracy can be considered a reliable evaluation metric.")
    else:
        print("The dataset is imbalanced.")
        print("Precision, Recall and F1-score should also be considered.")
```

```python
# ==========================================
# STEP 12 : Feature Engineering
# ==========================================

# Create Debt-to-Income Ratio
df["DebtToIncome"] = df["LoanAmount"] / df["Income"]

# Avoid division by zero
df["IncomePerYearEmployed"] = (
    df["Income"] /
    (df["EmploymentYears"] + 1)
)

print("=" * 60)
print("New Features Created")
print("=" * 60)

display(
    df[
        [
            "Income",
            "LoanAmount",
            "DebtToIncome",
            "EmploymentYears",
            "IncomePerYearEmployed"
        ]
    ].head()
)
```

```python
# ==========================================
# STEP 13 : Feature Scaling
# ==========================================

# Numerical columns to scale
numeric_features = [

    "Income",
    "CreditScore",
    "LoanAmount",
    "EmploymentYears",
    "DebtToIncome",
    "IncomePerYearEmployed"

]

# Create RobustScaler object
scaler = RobustScaler()

# Scale only numerical features
df[numeric_features] = scaler.fit_transform(df[numeric_features])

print("=" * 60)
print("Feature scaling completed.")
print("=" * 60)

display(df.head())
```

```python
# ==========================================
# STEP 14 : Final Dataset Check
# ==========================================

print("=" * 60)
print("First Five Rows")
print("=" * 60)

display(df.head())

print()

print("=" * 60)
print("Dataset Information")
print("=" * 60)

df.info()

print()

print("=" * 60)
print("Missing Values")
print("=" * 60)

print(df.isnull().sum())

print()

print("=" * 60)
print("Approved Column")
print("=" * 60)

print(df["Approved"].value_counts())
```

```python
# ==========================================
# STEP 15 : Save Clean Dataset
# ==========================================

# Save cleaned dataset
df.to_csv("clean_loan_dataset.csv", index=False)

print("=" * 60)
print("Dataset saved successfully.")
print("File Name: clean_loan_dataset.csv")
print("=" * 60)
```
