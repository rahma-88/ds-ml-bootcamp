# Car Price Prediction using Regression

**Assignment 4: Regression Models**

This notebook implements Linear Regression and Random Forest Regressor to predict car prices.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the cleaned dataset from Assignment 3
df = pd.read_csv('clean_car_dataset.csv')

print("Dataset loaded successfully!")
print(f"Shape: {df.shape}")
print(f"\nFirst few rows:")
print(df.head())
```

```python
# Prepare features and target variable
y = df['Price']
X = df.drop(columns=['Price', 'LogPrice'])

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"\nFeatures: {list(X.columns)}")
```

```python
# Split data into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

```python
# Train Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

print("Linear Regression model trained successfully!")
```

```python
# Train Random Forest Regressor model
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)
rf_model.fit(X_train, y_train)

print("Random Forest model trained successfully!")
```

```python
# Function to evaluate and display model performance
def evaluate_model(model, X_test, y_test, model_name):
    predictions = model.predict(X_test)
    
    r2 = r2_score(y_test, predictions)
    mae = mean_absolute_error(y_test, predictions)
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    
    print(f"\n{'='*50}")
    print(f"{model_name} Performance Metrics")
    print(f"{'='*50}")
    print(f"R² Score  : {r2:.4f}")
    print(f"MAE       : ${mae:,.2f}")
    print(f"MSE       : ${mse:,.2f}")
    print(f"RMSE      : ${rmse:,.2f}")
    print(f"{'='*50}")
    
    return {'r2': r2, 'mae': mae, 'mse': mse, 'rmse': rmse}

# Evaluate both models
lr_metrics = evaluate_model(lr_model, X_test, y_test, "Linear Regression")
rf_metrics = evaluate_model(rf_model, X_test, y_test, "Random Forest")
```

```python
# Sanity check: predict price for a single car
index = 0
sample = X_test.iloc[[index]]
actual_price = y_test.iloc[index]

lr_pred = lr_model.predict(sample)[0]
rf_pred = rf_model.predict(sample)[0]

print(f"\nSanity Check - Single Car Prediction:")
print(f"Actual Price              : ${actual_price:,.2f}")
print(f"Linear Regression Pred    : ${lr_pred:,.2f}")
print(f"Random Forest Pred        : ${rf_pred:,.2f}")
print(f"\nLinear Regression Error   : ${abs(actual_price - lr_pred):,.2f}")
print(f"Random Forest Error       : ${abs(actual_price - rf_pred):,.2f}")
```

```python
# Compare models
print("\nModel Comparison Summary:")
print(f"\nLinear Regression:")
print(f"  - Better for: Simple, interpretable predictions")
print(f"  - R² Score: {lr_metrics['r2']:.4f}")
print(f"  - RMSE: ${lr_metrics['rmse']:,.2f}")

print(f"\nRandom Forest:")
print(f"  - Better for: Capturing non-linear relationships")
print(f"  - R² Score: {rf_metrics['r2']:.4f}")
print(f"  - RMSE: ${rf_metrics['rmse']:,.2f}")

better_model = "Linear Regression" if lr_metrics['r2'] > rf_metrics['r2'] else "Random Forest"
print(f"\nBetter performing model: {better_model}")
```
