```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import (
    r2_score,
    mean_absolute_error,
    mean_squared_error
)
```

```python
df = pd.read_csv("clean_car_dataset1.csv")

df.head()
```

```python
print(df.shape)

print(df.info())
```

```python
print(df.columns.tolist())
```

```python
X = df.drop(columns=["Price", "LogPrice"])

y = df["Price"]

print(X.head())

print(y.head())
```

```python
# Step 3: Split Dataset
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training samples:", X_train.shape)

print("Testing samples:", X_test.shape)
```

```python
# Step 4: Train Linear Regression
linear_model = LinearRegression()

linear_model.fit(X_train, y_train)
```

```python
# Step 5: Train Random Forest Regression
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)
```

```python
# Step 6: Evaluate Models
def evaluate_model(model, X_test, y_test):

    prediction = model.predict(X_test)

    r2 = r2_score(y_test, prediction)

    mae = mean_absolute_error(y_test, prediction)

    mse = mean_squared_error(y_test, prediction)

    rmse = np.sqrt(mse)

    print("R²   :", round(r2,3))

    print("MAE  :", round(mae,2))

    print("MSE  :", round(mse,2))

    print("RMSE :", round(rmse,2))
```

```python
print("Linear Regression Performance")

evaluate_model(
    linear_model,
    X_test,
    y_test
)
```

```python
print("Random Forest Performance")

evaluate_model(
    rf_model,
    X_test,
    y_test
)
```

```python
# Step 7: Sanity Check

sample = X_test.iloc[[0]]

actual_price = y_test.iloc[0]

linear_prediction = linear_model.predict(sample)[0]

rf_prediction = rf_model.predict(sample)[0]

print("Actual Price:")

print(actual_price)

print()

print("Linear Regression Prediction:")

print(round(linear_prediction,2))

print()

print("Random Forest Prediction:")

print(round(rf_prediction,2))
```
