```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

DATA_PATH = "clean_car_dataset.csv"
df = pd.read_csv(DATA_PATH)
print(df.head())
```

```python
x = df.drop(columns=["Price","Log_price"])
y = df["Price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
#linear regression prediction
lnr = LinearRegression()
lnr.fit(x_train, y_train)
lnr_pred = lnr.predict(x_test)
print("linear regression")
print(lnr_pred)
```

```python
#random forest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)
print("randomforest")
print(rf_pred)
```

```python
#matrics
def print_metrics(name, y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    print(f"{name} performence: ")
    print(f"  r2  :  {r2:.3f}")
    print(f"  mae:   {mae:,.0f}")
    print(f"  mse:   {mse:,.0f}")
    print(f"  rmse:  {rmse:,.0f}")
```

```python
#Call it for both models.
print_metrics("Linear regression", y_test,lnr_pred)
print_metrics("random forest", y_test, rf_pred)
```

```python
#check
i = 5
one_x = x_test.iloc[[i]]
actual_price = y_test.iloc[i]

ln_predict = float(lnr.predict(one_x)[0])
rf_predict = float(rf.predict(one_x)[0])

print("\n one row prediction")
print(f"Actual price:            ${actual_price:,.0f}")
print(f"linear prediction:       ${ln_predict:,.0f}")
print(f"randomforest prediction: ${rf_predict:,.0f}")
```
