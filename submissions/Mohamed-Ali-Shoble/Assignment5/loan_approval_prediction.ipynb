```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import(accuracy_score, precision_score, recall_score, f1_score, confusion_matrix)
```

```python
# 1. Load Dataset
CSV_PATH=r"D:\Downloads\DS-ML COURSE\Jupiter(DS-ML)\clean_loan_dataset.csv"
df=pd.read_csv(CSV_PATH)
```

```python
# 2. Prepare Features & Target
x_features=df.drop(columns=["Approved"])
y_target=df["Approved"]
```

```python
# 3. Split Data
x_train, x_test, y_train, y_test=train_test_split(x_features, y_target, test_size=0.2, random_state=42, stratify=y_target)
print(f"Train size:{x_train.shape[0]} - Test size:{x_test.shape[0]}")
```

```python
# 4. Train Models
log_reg=LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(x_train, y_train)

random_frst=RandomForestClassifier(n_estimators=1000, random_state=42)
random_frst.fit(x_train, y_train)

d_tree=DecisionTreeClassifier(max_depth=5, random_state=42)
d_tree.fit(x_train, y_train)
```

```python
# 5. Evaluate Performance
def print_metrics(name,y_actual, y_pred, p_label=1):
    acc=accuracy_score(y_actual, y_pred)
    pre=precision_score(y_actual, y_pred, pos_label=p_label)
    re=recall_score(y_actual, y_pred, pos_label=p_label)
    f1=f1_score(y_actual, y_pred, pos_label=p_label)
    print(f"\n{name} matrics:")
    print(f"accuracy : {acc: .3f} (positive = Approved=1)")
    print(f"precision : {pre: .3f} (positive = Approved=1)")
    print(f"recall : {re: .3f} (positive = Approved=1)")
    print(f"F1 score : {f1: .3f} (positive = Approved=1)")

def conf_matrix(name,y_actual,y_pred):
    cfs_matrix=confusion_matrix(y_actual,y_pred,labels=[1,0])
    cfs_matrix_df=pd.DataFrame(cfs_matrix, index=["Actual: Approved(1)", "Actual: Rejected(0)"], 
                               columns=["Predicted: Approved(1)", "Predicted: Rejected(0)"])
    print(f"\n{name} confusion matrix:")
    print(cfs_matrix_df)

def label_str(x):
    return "Approved" if x==1 else "Rejected"

log_reg_pred=log_reg.predict(x_test)
print_metrics("Logistic Regression",y_test,log_reg_pred,p_label=1)
conf_matrix("Logistic Regression", y_test,log_reg_pred)

rf_predictor=random_frst.predict(x_test)
print_metrics("Random Forest",y_test,rf_predictor,p_label=1)
conf_matrix("Rnadom Forest", y_test,rf_predictor)

dtree_predictor=d_tree.predict(x_test)
print_metrics("Decision Tree",y_test,dtree_predictor,p_label=1)
conf_matrix("Decision Tree", y_test,dtree_predictor)
```

```python
# 6. Sanity Check
i=4
x_one_df=x_test.iloc[[i]]
y_one_df=y_test.iloc[i]
print("Sanity check:")
print("Actual Label: ",y_one_df)
print("Logistic Regression: ", label_str(int(log_reg.predict(x_one_df)[0])))
print("Random Forest: ", label_str(int(random_frst.predict(x_one_df)[0])))
print("Decision Tree: ", label_str(int(d_tree.predict(x_one_df)[0])))
```
