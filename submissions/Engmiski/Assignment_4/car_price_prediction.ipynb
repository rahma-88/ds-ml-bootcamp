```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import (
    r2_score,
    mean_absolute_error,
    mean_squared_error
)
df = pd.read_csv("Data/car_l3_clean_ready.csv")

df.head()
```

```python
print(df.shape)
```

```python
X = df.drop(columns=["Price", "LogPrice"])

y = df["Price"]

print(X.head())

print(y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
linear_model = LinearRegression()

linear_model.fit(X_train, y_train)
```

```python
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)
```

```python
def evaluate_model(model, X_test, y_test):

    prediction = model.predict(X_test)

    r2 = r2_score(y_test, prediction)

    mae = mean_absolute_error(y_test, prediction)

    mse = mean_squared_error(y_test, prediction)

    rmse = np.sqrt(mse)

    print(f"R²   : {r2:.4f}")
    print(f"MAE  : {mae:.2f}")
    print(f"MSE  : {mse:.2f}")
    print(f"RMSE : {rmse:.2f}")

    return prediction
```

```python
print("Linear Regression Performance")

linear_prediction = evaluate_model(
    linear_model,
    X_test,
    y_test
)
```

```python
print("Random Forest Performance")

rf_prediction = evaluate_model(
    rf_model,
    X_test,
    y_test
)
```

```python
i = 0

sample = X_test.iloc[[i]]

actual_price = y_test.iloc[i]

linear_price = linear_model.predict(sample)[0]

rf_price = rf_model.predict(sample)[0]

print("Actual Price:", actual_price)

print("Linear Regression Prediction:", linear_price)

print("Random Forest Prediction:", rf_price)
```
