```python
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.preprocessing import StandardScaler
```

```python
# read csv file

CSV_PATH = "car_clean_dataset.csv"

df = pd.read_csv(CSV_PATH)
```

```python
# SHOW HEAD DATA

df.head()
```

```python
# split features

x = df.drop(columns=["Price", "Log_Price"])
y = df["Price"]
```

```python
# train/test split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# train linear regression

lr = LinearRegression()

lr.fit(x_train, y_train)
lr_predict = lr.predict(x_test)

print("========================================\n===== Linear Regression Prediction =====\n========================================")
print(lr_predict)
```

```python
rfr = RandomForestRegressor(n_estimators=100, random_state=42)

rfr.fit(x_train, y_train)
rfr_predict = rfr.predict(x_test)

print("===============================================\n===== Random Forest Regression Prediction =====\n===============================================")
print(rfr_predict)
```

```python
# helper function

def logRegressionMetrics(model_name, y_true, y_pred):
    """Prints standard regression evaluation metrics."""

    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)

    r2_percentage = r2*100

    print(f"\n=================================\n=== {model_name} Performance ===\n=================================")
    print(f"Mean Absolute Error: {mae:,.4f}")
    print(f"Mean Squared Error: {mse:,.4f}")
    print(f"Root Mean Squared Error: {rmse:,.4f}")
    
    print(f"R² Score:           {r2:.4f} ( {r2_percentage:.2f}% )\n")


logRegressionMetrics("Linear Regression", y_test, lr_predict)
logRegressionMetrics("Random Forest", y_test, rfr_predict)
```

```python
# feature scalling

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train) # fits only on train
x_test_scaled = scaler.transform(x_test) # uses train weight on test


# Convert back to DataFrame just for a clean print
x_train_scaled_df = pd.DataFrame(x_train_scaled, columns=x_train.columns)

# Print the first few rows to see the values centered around 0
print("===================================\n=== First 5 Rows of Scaled Data ===\n===================================")
print(x_train_scaled_df.head())
```

# Scaler Learned 

```python
# 2. Print the learned mean and scale (standard deviation) for each feature
print("=================================\n=== Scaler Learned Parameters ===\n=================================")
for feature, mean, std in zip(x_train.columns, scaler.mean_, scaler.scale_):
    print(f"Feature: {feature:<20} | Mean: {mean:.4f} | Std Dev: {std:.4f}")
```

# Train and Test Score Models

```python
# train score
print("===============================\n=== Linear Regression SCORE ===\n===============================")
lr_train_score = lr.score(x_train, y_train)
lr_test_score = lr.score(x_test, y_test)
print(f"Train Score :- {lr_train_score:.1f}"  )
print(f"Test Score :- {lr_test_score:.1f}" )
```

```python

print("===========================\n=== Random Forest SCORE ===\n===========================")
rfr_train_score = rfr.score(x_train, y_train)
rfr_test_score = rfr.score(x_test, y_test)
print(f"Train Score :- {rfr_train_score:.1f}"  )
print(f"Test Score :- {rfr_test_score:.1f}" )
```
