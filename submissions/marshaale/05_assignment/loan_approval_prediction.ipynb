```python
import pandas as pd 
import numpy as np
from pathlib import Path
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score,confusion_matrix
```

```python
DATA_DIR=Path().cwd() / "data"
TRAIN_CSV_FILE = DATA_DIR / "train.csv"
TEST_CSV_FILE = DATA_DIR / "test.csv"
RANDOM_STATE = 42
print(DATA_DIR)
print(TRAIN_CSV_FILE)
print(TEST_CSV_FILE)
```

## Load Dataset

```python
train_df = pd.read_csv(TRAIN_CSV_FILE)
test_df = pd.read_csv(TEST_CSV_FILE)
```

## Prepare Features & Target

```python
X_train = train_df.drop(columns=['Status'])
y_train = train_df['Status']

X_test = test_df.drop(columns=['Status'])
y_test = test_df['Status']
```

## Train Models (Reproduce Lesson)

```python
rfc = RandomForestClassifier(n_estimators=1000,random_state=RANDOM_STATE)
log_reg = LogisticRegression(max_iter=1000)
svm = SVC(max_iter=1000,random_state=RANDOM_STATE,kernel='sigmoid')
dtree = DecisionTreeClassifier(random_state=RANDOM_STATE)
```

```python
dtree.fit(X_train,y_train)
```

```python
rfc_predicts = rfc.predict(X_test)
log_reg_predicts = log_reg.predict(X_test)
svm_predicts = svm.predict(X_test)
dtree_predicts = dtree.predict(X_test)
```

## Evaluate Performance

```python
# Helpers

def print_metrics(name:str, y_true:pd.Series, y_pred:np.ndarray, pos_label=1):
    acc = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, pos_label=pos_label)
    rec = recall_score(y_true, y_pred, pos_label=pos_label)
    f1 = f1_score(y_true, y_pred, pos_label=pos_label)
    
    print(f"\n{name} Performance:")
    print(f"  Accuracy : {acc:.3f}")
    print(f"  Precision: {precision:.3f}")
    print(f"  Recall   : {rec:.3f}")
    print(f"  F1-Score : {f1:.3f}")

def print_confusion_metrics(y_true:pd.Series, y_pred:np.ndarray):
    cm = confusion_matrix(y_true,y_pred,labels=[1,0])
    cm_df = pd.DataFrame(cm,
        index=["Actual: Approved (1)", "Actual: Rejected (0)"],
        columns=["Pred: Approved (1)", "Pred: Rejected (0)"],
    )
    print(cm)
    return cm_df

def label_str(v:int|float):
    return "Approved (1)" if v == 1 else "Rejected (0)"
```

```python
print_metrics("Logistic Regression Classifier",y_test,log_reg_predicts)
print_metrics("Random Forest Classifier",y_test,rfc_predicts)
print_metrics("SVM Classifier",y_test,svm_predicts)
print_metrics("Decision Tree Classifier",y_test,dtree_predicts)
```

```python
print_confusion_metrics(y_test,log_reg_predicts)
```

```python
print_confusion_metrics(y_test,rfc_predicts)
```

```python
print_confusion_metrics(y_test,dtree_predicts)
```

```python
print_confusion_metrics(y_test,svm_predicts)
```
