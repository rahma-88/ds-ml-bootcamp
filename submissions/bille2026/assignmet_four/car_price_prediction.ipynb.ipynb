```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
#  1) load the cleanin dataset 

CSV_path = "clean_raw_car_dataset.csv"
df = pd.read_csv(CSV_path)
```

```python
print(df.head())
```

```python
# 2) split features  (x) and target (y) 

#  we predict price we also drop logprice so we don't leak target info

x = df.drop(columns=["Price" , "LogPrice"])
y = df["Price"]
```

```python
#  3) train / test split for fair evaluation
 
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size=0.2 , random_state=42)
```

```python

# 4) train linear regression

lr= LinearRegression()
lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)
```

```python
print("linear regression predictions ")
print(lr_pred)
```

```python
rf = RandomForestRegressor(n_estimators=100 , random_state=42)
rf.fit(x_train, y_train)
rf_pred =rf.predict(x_test)
```

```python
print("random forest predictions ")
print(rf_pred)
```

```python
# 6 ) helper to print metrics 

def  print_metrics(name , y_true , y_pred):
    r2= r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true , y_pred)
    mse = mean_squared_error(y_true , y_pred)
    rmse = np.sqrt(mse)

    print(f"\n{name} performence:")
    print(f" r2 : {r2:.3f}")
    print(f" MAE: {mae:.0f}")
    print(f" MSE: {mse:.0f}")
    print(f" RMSE: {rmse:.0f}")
```

```python
 # 7 ) show results for both models 

print_metrics(LinearRegression , y_test , lr_pred)
print_metrics(RandomForestRegressor, y_test,rf_pred)
```

```python
# 8) single_raw prediction [sanity check]

i =3 
x_one_df = x_test.iloc[[i]]
y_true = y_test.iloc[i]
p_lr_one = float(lr.predict(x_one_df)[0])
p_rf_one = float(rf.predict(x_one_df)[0])

print ("\n single_raw prediction (raw index 3): ")
print(f" actual price: ${y_true:,.0f}")

print(f"linear regression prediction: ${p_lr_one:,.0f} ")
print(f"random forest predictions: ${p_rf_one:,.0f} ")
```
