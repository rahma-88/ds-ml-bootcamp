```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
CSv_PATH="car_dataset_cleaned.csv"
df=pd.read_csv(CSv_PATH)
print(df.shape)
```

```python
#classify y and x
x=df.drop(columns=["Price","Logprice"])
y=df["Price"]
print(y)
```

```python
#spliting the data into train and test
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
#prredicting y using linearregression
lr=LinearRegression()
lr.fit(x_train,y_train)
lr_pred=lr.predict(x_test)
# using ramdomforest
rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(x_train,y_train)
rf_pred=rf.predict(x_test)
```

```python
#figuering how our model close or far
def print_metrics(name,y,y_predict):
    r2=r2_score(y,y_predict)
    mae=mean_absolute_error(y,y_predict)
    mse=mean_squared_error(y,y_predict)
    print(f"prediction of:{name}")
    rmse=np.sqrt(mse)
    print(f"r2:{r2:.3f}")
    print(f"mae:{mae:.0f}")
    print(f"mse:{mse:.0f}")
    print(f"rmse:{rmse:.0f}")
print_metrics("linear",y_test,lr_pred)
print_metrics("randomforest",y_test,rf_pred)

    
```

```python
comparison_df = pd.DataFrame({
    'Actual Price': y_test,
    'Linear Regression Prediction': lr_pred,
    'Random Forest Prediction': rf_pred
})
comparison_df.head(17)
```

```python
print("Training R²:", lr.score(x_train, y_train))
print("Testing R²:", lr.score(x_test, y_test))
print("RandomForest:")
print("Training R²:", rf.score(x_train, y_train))
print("Testing R²:", rf.score(x_test, y_test))

#knowing weather our model over fitting or undervitting 
```
