```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import (
    r2_score,
    mean_absolute_error,
    mean_squared_error
)
```

```python
# 1: Load dataset
df = pd.read_csv("clean_car_dataset.csv")

# Display the first 5 rows
print(df.head())

# Display dataset shape
print("\nDataset Shape:")
print(df.shape)
```

```python
# 2: Features & Target
# Target variable
y = df["Price"]

# Feature variables
X = df.drop(columns=["Price", "LogPrice"])

# Display feature names
print("Features:")
print(X.columns.tolist())

print("\nTarget:")
print(y.name)
```

```python
# 3: Splitting Train & Test

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)
```

```python
# 4: Train Models
# Training Linear Regression

linear_model = LinearRegression()

linear_model.fit(X_train, y_train)

print("Linear Regression model trained successfully.")
```

Train Random Forest

```python
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)

print("Random Forest model trained successfully.")
```

Helper Function for Evaluation

```python
def evaluate_model(model, X_test, y_test, model_name):

    predictions = model.predict(X_test)

    r2 = r2_score(y_test, predictions)
    mae = mean_absolute_error(y_test, predictions)
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)

    print("=" * 40)
    print(model_name)
    print("=" * 40)

    print(f"R²   : {r2:.2f}")
    print(f"MAE  : {mae:.2f}")
    print(f"MSE  : {mse:.2f}")
    print(f"RMSE : {rmse:.2f}")
```

Evaluate Both Models

```python
evaluate_model(
    linear_model,
    X_test,
    y_test,
    "Linear Regression Performance"
)

print()

evaluate_model(
    rf_model,
    X_test,
    y_test,
    "Random Forest Performance"
)
```

Sanity Check

```python
# Select one sample from the test set
sample = X_test.iloc[[0]]

# Actual price
actual_price = y_test.iloc[0]

# Predictions
linear_prediction = linear_model.predict(sample)[0]
rf_prediction = rf_model.predict(sample)[0]

print("Actual Price:", actual_price)
print("Linear Regression Prediction:", linear_prediction)
print("Random Forest Prediction:", rf_prediction)
```
