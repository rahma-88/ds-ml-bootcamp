```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

from sklearn.metrics import confusion_matrix
```

```python
# Load the cleaned loan dataset created in Part B1

df = pd.read_csv("clean_loan_dataset.csv")

print("=" * 60)
print("STEP 1: LOAD DATASET")
print("=" * 60)

print(df.head())

print("\nDataset Shape:")
print(df.shape)
```

```python
# Separate the target variable (Approved) from the feature variables

X = df.drop(columns=["Approved"])
y = df["Approved"]

print("=" * 60)
print("STEP 2: PREPARE FEATURES & TARGET")
print("=" * 60)

print("Features Shape:", X.shape)
print("Target Shape:", y.shape)

print("\nFeature Columns:")
print(X.columns.tolist())
```

```python
# Split the dataset into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

print("=" * 60)
print("STEP 3: TRAIN TEST SPLIT")
print("=" * 60)

print("Training Set:", X_train.shape)
print("Testing Set :", X_test.shape)
```

```python
# Train the Logistic Regression model

log_model = LogisticRegression(
    max_iter=1000,
    random_state=42
)

log_model.fit(X_train, y_train)

print("=" * 60)
print("STEP 4: LOGISTIC REGRESSION")
print("=" * 60)

print("Logistic Regression model trained successfully.")
```

```python
print(X_train.isnull().sum())
print("\nTotal NaNs:", X_train.isnull().sum().sum())
```

```python
X_train = X_train.fillna(X_train.median(numeric_only=True))
X_test = X_test.fillna(X_test.median(numeric_only=True))
```

```python
binary_cols = ["HasCollateral", "PreviousDefaults"]

for col in binary_cols:
    X_train[col] = X_train[col].fillna(X_train[col].mode()[0])
    X_test[col] = X_test[col].fillna(X_test[col].mode()[0])
```

```python
log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train, y_train)

print("Logistic Regression trained successfully")
```
