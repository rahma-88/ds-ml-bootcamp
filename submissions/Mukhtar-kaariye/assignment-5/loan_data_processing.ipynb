```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("dataset/raw_loan_dataset.csv")
```

```python
print("=" * 60)
print("STEP 1: LOAD & INSPECT")
print("=" * 60)

print("\nFirst 10 Rows:")
print(df.head(10))
```

```python
print("\nDataset Shape:")
print(df.shape)
```

```python
print("\nDataset Information:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nHasCollateral Value Counts:")
print(df["HasCollateral"].value_counts(dropna=False))

print("\nPreviousDefaults Value Counts:")
print(df["PreviousDefaults"].value_counts(dropna=False))

print("\nApproved Value Counts:")
print(df["Approved"].value_counts(dropna=False))
```

```python
print("\nKey Issues Found:")
print("- Income and LoanAmount contain currency symbols.")
print("- Yes/No columns have inconsistent spellings.")
print("- Approved contains values other than Yes and No.")
print("- Dataset contains missing values.")
print("- Possible duplicate rows.")
```

```python
print("=" * 60)
print("STEP 2: CLEAN CURRENCY FORMATTING")
print("=" * 60)

# Remove $ and commas from Income
df["Income"] = (
    df["Income"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

# Remove $ and commas from LoanAmount
df["LoanAmount"] = (
    df["LoanAmount"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

# Convert to numeric
df["Income"] = pd.to_numeric(df["Income"], errors="coerce")
df["LoanAmount"] = pd.to_numeric(df["LoanAmount"], errors="coerce")
```

```python
print("\nData Types:")
print(df[["Income", "LoanAmount"]].dtypes)

print("\nIncome Summary:")
print(df["Income"].describe())

print("\nLoanAmount Summary:")
print(df["LoanAmount"].describe())
```

```python
print("=" * 60)
print("STEP 3: FIX CATEGORY ERRORS")
print("=" * 60)

# Normalize HasCollateral
df["HasCollateral"] = (
    df["HasCollateral"]
    .astype(str)
    .str.strip()
    .str.lower()
    .replace({
        "yes": "Yes",
        "y": "Yes",
        "no": "No",
        "n": "No",
        "nan": np.nan
    })
)

# Normalize PreviousDefaults
df["PreviousDefaults"] = (
    df["PreviousDefaults"]
    .astype(str)
    .str.strip()
    .str.lower()
    .replace({
        "yes": "Yes",
        "y": "Yes",
        "no": "No",
        "n": "No",
        "nan": np.nan
    })
)

# Normalize Approved
df["Approved"] = (
    df["Approved"]
    .astype(str)
    .str.strip()
    .str.lower()
    .replace({
        "yes": "Yes",
        "approved": "Yes",
        "y": "Yes",
        "no": "No",
        "rejected": "No",
        "n": "No",
        "nan": np.nan
    })
)
```

```python
print("\nHasCollateral:")
print(df["HasCollateral"].value_counts(dropna=False))

print("\nPreviousDefaults:")
print(df["PreviousDefaults"].value_counts(dropna=False))

print("\nApproved:")
print(df["Approved"].value_counts(dropna=False))
```

```python
print("=" * 60)
print("STEP 4: IMPUTE MISSING VALUES")
print("=" * 60)

# Fill missing numeric values with median
df["Income"] = df["Income"].fillna(df["Income"].median())
df["CreditScore"] = df["CreditScore"].fillna(df["CreditScore"].median())
df["LoanAmount"] = df["LoanAmount"].fillna(df["LoanAmount"].median())
df["EmploymentYears"] = df["EmploymentYears"].fillna(df["EmploymentYears"].median())

# Fill missing categorical values with mode
df["HasCollateral"] = df["HasCollateral"].fillna(df["HasCollateral"].mode()[0])
df["PreviousDefaults"] = df["PreviousDefaults"].fillna(df["PreviousDefaults"].mode()[0])
df["Approved"] = df["Approved"].fillna(df["Approved"].mode()[0])
```

```python
print("\nMissing Values After Imputation:")
print(df.isnull().sum())
```

```python
print("=" * 60)
print("STEP 5: REMOVE DUPLICATES")
print("=" * 60)

# Shape before removing duplicates
print("Shape Before:", df.shape)

# Count duplicates
duplicates = df.duplicated().sum()
print("Duplicate Rows:", duplicates)

# Remove duplicates
df = df.drop_duplicates()

# Shape after removing duplicates
print("Shape After:", df.shape)
```

```python
print("=" * 60)
print("STEP 6: OUTLIERS (IQR CAPPING)")
print("=" * 60)

# Columns to cap
columns = ["Income", "CreditScore", "LoanAmount", "EmploymentYears"]

for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower=lower, upper=upper)

    print(f"\n{col}")
    print(f"Lower Bound: {lower:.2f}")
    print(f"Upper Bound: {upper:.2f}")
```

```python
print("\nSummary After IQR Capping:")
print(df[columns].describe())
```

```python
print("=" * 60)
print("STEP 7: LABEL ENCODING")
print("=" * 60)

# Encode Yes/No columns
mapping = {"Yes": 1, "No": 0}

df["Approved"] = df["Approved"].map(mapping)
df["HasCollateral"] = df["HasCollateral"].map(mapping)
df["PreviousDefaults"] = df["PreviousDefaults"].map(mapping)
```

```python
print("\nApproved Class Distribution:")
print(df["Approved"].value_counts())

print("\nHasCollateral Distribution:")
print(df["HasCollateral"].value_counts())

print("\nPreviousDefaults Distribution:")
print(df["PreviousDefaults"].value_counts())

print("\nData Types:")
print(df[["Approved", "HasCollateral", "PreviousDefaults"]].dtypes)
```

```python
print("=" * 60)
print("STEP 8: CLASS BALANCE CHECK")
print("=" * 60)

print("\nApproved vs Rejected Counts:")
print(df["Approved"].value_counts())

print("\nApproved vs Rejected Proportions:")
print(df["Approved"].value_counts(normalize=True))
```

### Class Balance Observation

The class distribution shows that both Approved (1) and Rejected (0) are reasonably represented in the dataset. The classes are fairly balanced, so Accuracy can be used as an evaluation metric. However, Precision, Recall, and F1-Score should also be considered to obtain a more complete assessment of the classification models.

```python
print("=" * 60)
print("STEP 9: FEATURE ENGINEERING")
print("=" * 60)

# Debt-to-Income Ratio
df["DebtToIncome"] = df["LoanAmount"] / (df["Income"] + 1)

# Income Per Year Employed
df["IncomePerYearEmployed"] = df["Income"] / (df["EmploymentYears"] + 1)
```

```python
print("\nNew Features Added:")
print(df[["DebtToIncome", "IncomePerYearEmployed"]].head())

print("\nDataset Shape:")
print(df.shape)
```

```python
print("=" * 60)
print("STEP 10: FEATURE SCALING")
print("=" * 60)

# Numeric columns only
numeric_columns = [
    "Income",
    "CreditScore",
    "LoanAmount",
    "EmploymentYears",
    "DebtToIncome",
    "IncomePerYearEmployed"
]

# Create scaler
scaler = MinMaxScaler()

# Scale the columns
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

print("\nScaled Features (First 5 Rows):")
print(df[numeric_columns].head())
```

```python
print("\nMinimum Values:")
print(df[numeric_columns].min())

print("\nMaximum Values:")
print(df[numeric_columns].max())

print("\nLabel Encoded Columns (Not Scaled):")
print(df[["Approved", "HasCollateral", "PreviousDefaults"]].head())
```

### Feature Scaling

I selected **MinMaxScaler** from `sklearn.preprocessing` because it scales numeric features into a range between 0 and 1. This is useful when the features have different scales, such as Income and LoanAmount, while preserving the relationship between values.

The label-encoded columns (**Approved**, **HasCollateral**, and **PreviousDefaults**) were not scaled because they already represent binary categories (0 and 1).

Scikit-learn Developers. *MinMaxScaler*. https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html
